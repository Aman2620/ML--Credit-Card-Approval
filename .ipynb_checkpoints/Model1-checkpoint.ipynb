```python
import datetime
from datetime import timedelta
```

```python
def convert_date(total_days):
    real_date = datetime.date.today()
    date = (real_date + timedelta(days = total_days)).strftime('%d-%m-%Y')
    return date
```

```python
main_df['DAYS_BIRTH'] = main_df['DAYS_BIRTH'].apply(convert_date)
```

```python
main_df.head()
```

```python
#Lets do the same for DAYS_Employed
main_df['DAYS_EMPLOYED'] = main_df['DAYS_EMPLOYED'].apply(convert_date)
```

```python
main_df.head()
```

```python
#Now let's convert the DAYS_BIRTH into Age of the applicants
def age_applicants(birth_date):
    today = datetime.date.today()
    dateofbirth=datetime.datetime.strptime(birth_date,'%d-%m-%Y')
    age = today.year - dateofbirth.year
    return age
```

```python
main_df['DAYS_BIRTH']=main_df['DAYS_BIRTH'].apply(age_applicants)
```

```python
main_df.rename(columns={'DAYS_BIRTH':'Age'},inplace=True)
```

<b>
    Now according to the explanation given for the status column, there are certain assumptions we can draw some conclusions<br>
        1. Values 2,3,4,5 are those people who do not pay on time or those who have bad credits or debts.<br>
        2. Value 1 are those people who have very minimal risk, therefore we can ignore it<br>
        3. Vaules C,X are those people who have paid their credit or those who have no loan, therefore not at all risky<br>
    So what we can do is create another column by considering the above points
</b>


```python
#Lets also convert C & X from Status column into numerical
main_df.loc[main_df['STATUS']=='C','STATUS']=6
main_df.loc[main_df['STATUS']=='X','STATUS']=7
```

```python
main_df['STATUS']=main_df['STATUS'].astype(float)
```

```python
def solved(state):
    target=''
    if state in (2,3,4,5):
        target=1 #Risky
    else :
        target=0 #not Risky
        
    return target
```

```python
main_df['Target']=main_df['STATUS'].apply(solved)
```

```python
#The target variable is therefore unbalanced, hence let's use over samppling
from imblearn.combine import SMOTETomek
```

```python
smk=SMOTETomek(random_state=12)
```

```python
x_res,y_res=smk.fit_resample(x,y)
```

```python
x_res.shape
```

```python
y_res.shape
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import ConfusionMatrixDisplay,accuracy_score
```

```python
model_list=[LogisticRegression,RandomForestClassifier,DecisionTreeClassifier,KNeighborsClassifier]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_res,y_res,test_size=0.2,random_state=567)
```

```python
accuracy =[]
for model in model_list :
    model = model()
    model.fit(x_train,y_train)
    y_pred = model.predict(x_test)
    ConfusionMatrixDisplay.from_predictions(y_test,y_pred)
    accuracy.append(accuracy_score(y_test,y_pred))
```
